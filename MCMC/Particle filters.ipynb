In the lectures, we saw that the importance weights in SMC can be written as

\begin{equation}
w_n = \frac{p(y_{1:n}|x_{1:n})p(x_{1:n})}{q(x_{1:n})},
\end{equation}
where $x_{1:n} \sim q(x_{1:n}).$

However, as we discussed, it is difficult to sample directly from $q(x_{1:n})$, and difficult to evaluate $w_n$ (when in the form above). Luckily we can make some simplifying assumptions by observing that we are dealing with a hidden Markov model (HMM). Your task is to show, based on these simplifying assumptions that, 
$$w_n = \frac{p(y_{1:n}|x_{1:n})p(x_{1:n})}{q(x_{1:n})}=\alpha(x_{n})w_{n-1},$$
where
$$x_{n}\sim q(x_n|x_{n-1})$$
$$\alpha(x_{n}) =\frac{p(y_{n}|x_{n})p(x_{n}|x_{n-1})}{q(x_{n}|x_{n-1})}$$
$$w_{n-1}=\frac{p(y_{1:n-1}|x_{1:n-1})p(x_{1:n-1})}{q(x_{1:n-1})}$$

# ***2. The Stochastic Volatility Model***
In this assignment the stochastic volatililty (SV) model was used:

\begin{equation*}
        x_1 \sim \mathcal{N}(0, \frac{\sigma^2}{1 - \alpha ^2}),
\end{equation*}

\begin{equation*}
        x_n \sim \mathcal{N}(\alpha x_{n-1}, \sigma^2),
\end{equation*}

\begin{equation*}
        y_n \sim \mathcal{N}(0,  \beta^2\exp(x_{n})),
\end{equation*}

for $n=1,...,T$.

Let $T=100$, $\beta=0.64$, $\alpha=0.91$ and $\sigma=1$.

# ***3. Sequential Importance Sampling***

\begin{align*}
\mathbb{E}_{p(x_n|y_{1:n})}[x_n] 
&= \int_{x_n} p(x_n|y_{1:n}) x_n dx_n 
\approx \int_{x_n} \sum_k \tilde{w}^k_n \delta_{x^k_n}(x_n) x_n dx_n = \sum_k \tilde{w}^k_n x^k_n
\end{align*}

For simplicity, let the proposal distribution $q(x_n|x_{n-1}) = p(x_n|x_{n-1})$. 

\begin{align*}
\alpha (x^k_n) 
= \frac{p(y_n|x^k_n) p(x^k_n|x^k_{n-1})}{q(x^k_n|x^k_{n-1})} = p(y_n|x^k_n) = \mathcal{N}(y_n | 0,  \beta^2\exp(x^k_{n})) 
\end{align*}
Since $q(x_n|x_{n-1}) = p(x_n|x_{n-1})$.

\begin{align*}
x^k_n & \sim q(x^k_n|x^k_{n-1}) = p(x^k_n|x^k_{n-1}) = \mathcal{N}(\alpha x^k_{n-1}, \sigma^2) \\
y_n & \sim p(y_n|x^k_n) = \mathcal{N}(0,  \beta^2\exp(x^k_{n}))\\
w^k_n & = \mathcal{N}(y_n | 0,  \beta^2\exp(x^k_{n})) w^k_{n-1}
\end{align*}

Computation of MSE
\begin{align*}
\hat{x}_n &= \mathbb{E}_{p(x_n|y_{1:n})}[x_n] = \sum_k \tilde{w}^k_n x^k_n\\
\text{MSE} &= \frac{1}{T}\sum_n (\hat{x}_n - x_n)^2
\end{align*}

```python
# Implement BPF here
K = 100
T = 100
beta = 0.64
alpha = 0.91
sigma2 = 1

def pyx(yn, xkn, beta = beta): # alpha(x^k_n) 
    return norm.pdf(yn, 0, beta * np.sqrt(np.exp(xkn)))

def sample_yn(xkn, beta = beta):
    return norm.rvs(0, beta * np.sqrt(np.exp(xkn)))

def sample_xn(xkn_1, alpha = alpha, sigma2 = sigma2):
    return norm.rvs(alpha * xkn_1, np.sqrt(sigma2))

# resampling schemes in question 4
def multnorm_resamp(K, w):
    # return np.random.multinomial(K, w)
    mulinomial = np.random.multinomial(K, w)
    resamp = np.zeros(K, dtype=int)
    i = 0
    for j in range(0, K):
        k = i + mulinomial[j]
        while i < k:
            resamp[i] = j
            i += 1
    return resamp

def strat_resamp(K, w):
    a = np.zeros(K+2)
    for i in range(1,K+1):
        a[i] = a[i-1] + w[i-1]
    a[K+1] = 1
    v = np.random.uniform(0, 1/K, K) + 1/K * np.arange(0,K)
    return np.searchsorted(a, v) - 1

def noresampling(K, w):
    return np.arange(0, K)
```

```python
class BFP:
    def __init__(self, scheme="None", K=K, T=T, alpha = 0.91, beta = 0.64, sigma2 = 1, inference = False, obs = []) -> None:
        self.K = K
        self.xn = np.zeros(T)
        self.xkn = np.zeros((T, K))
        self.wn = np.zeros((T, K))
        self.yn = np.zeros(T)
        if inference:
            self.yn = npa(obs)
        self.w_tilde = np.zeros((T, K))
        self.e_xn = np.zeros(T)
        self.samp = noresampling
        self.scheme = "No resampling"

        if scheme=='multnorm':
            self.samp = multnorm_resamp
            self.scheme = "Multinormial resampling"
        elif scheme=='stratified':
            self.samp = strat_resamp
            self.scheme = "Stratified resamplinig"

        self.xn[0] = np.random.normal(0, np.sqrt(sigma2/(1-alpha**2)))
        if not inference:
            self.yn[0] = sample_yn(self.xn[0], beta)

        self.xkn[0] = npa([sample_xn(self.xn[0], alpha, sigma2) for i in range(0,K)])
        self.wn[0] = npa([pyx(self.yn[0], self.xkn[0][i], beta) for i in range(0,K)])
        self.w_tilde[0] = self.wn[0]/np.sum(self.wn[0])

        self.log_likelihood = 0
        logK = np.log(K)

        for n in range(1, T):
            # resampling step
            ak = self.samp(K, self.w_tilde[n-1])
            self.xn[n] = sample_xn(self.xn[n-1], alpha, sigma2)
            if not inference:
                self.yn[n] = sample_yn(self.xn[n], beta)

            self.xkn[n] = npa([sample_xn(self.xkn[n-1][ak[i]], alpha, sigma2) for i in range(0,K)])

            pyxk = npa([pyx(self.yn[n], self.xkn[n][i], beta) for i in range(0,K)])
            if inference:
                self.log_likelihood += np.log(np.sum(pyxk)) - logK
            # self.wn[n] = pyxk * self.wn[n-1][ak]

            if scheme == "None":
                self.wn[n] = pyxk * self.wn[n-1]
            else:
                self.wn[n] = pyxk

            self.w_tilde[n] = self.wn[n]/np.sum(self.wn[n])

            self.e_xn[n] = np.sum(self.w_tilde[n] * self.xkn[n])
            if scheme != "None":
                new_xkn = np.zeros((n, K))
                for i in range(0, K):
                    new_xkn[:, i] = self.xkn[:n, ak[i]]
                self.xkn[:n, :] = new_xkn

    
    def estimate(self):
        return self.e_xn
```

```python
bfp5 = BFP(K=5)
bfp100 = BFP(K=100)
bfp500 = BFP(K=500)
```

```python
# 2. histogram
fig, axs = plt.subplots(1, 4, figsize=(13, 5))
fig.suptitle('Distribution of normalized particle weights (K = 100)')

npoints = (2, 5, 50, 100)

for i in range(0, 4):
    axs[i].hist(bfp100.w_tilde[npoints[i]-1], 20, align='left')
    axs[i].set(xlabel = 'Importance weights')
    axs[i].set_title('n={}'.format(npoints[i]))

fig.tight_layout(pad=1)
plt.show()
```

```python
# 3 trajectory

for i in range(0, 100):
    plt.plot(bfp100.xkn[:,i])
plt.plot(bfp100.xn, c='blue', marker='*', ls='-', label='True Volatility')
plt.legend()
plt.xlabel('n')
plt.ylabel(r'$\hat{w}$')
plt.title('Inferred x1:T using K = 100 particles and SIS')
plt.show()
```

```python
def mse(bfp):
    e_xn = bfp.estimate()
    mse = np.mean((e_xn - bfp.xn) ** 2)
    return mse

print('MSE for K = 5 is ' + str(mse(bfp5)))
print('MSE for K = 100 is ' + str(mse(bfp100)))
print('MSE for K = 500 is ' + str(mse(bfp500)))
```

The MSE is usually less when K is more. 
The reason might be that if we use more particles, the estimation will be better. Therefore the mean square error is less.

If $K \rightarrow \infty$ the estimation will be the exact value.

# ***5. Parameter Inference using the Marginal Likelihood***


$$
p(y_{1:T}|\theta) \approx \prod^n_{i=1} \sum_k \frac{1}{k} p(y_i|x^k_i, \theta)
$$

Coarse grid (say, $8\times8\times 8$) for $\theta$ in the interval ($0$, $1$).

```python
def param_inference(obs, alpha, beta, sigma2):
    b = BFP("multnorm", K = 100, T = obs.size, alpha = alpha, beta = beta, sigma2 = sigma2, inference=True, obs=obs)
    return b.log_likelihood  
```

```python
# load observations.npy from Canvas and do the grid search below
obs = np.load("observations.npy")
N = obs.size

lsa = np.linspace(0.1, 0.9, 8)
lsb = np.linspace(0.1, 1, 8)
lss = np.linspace(0.1, 1, 8) ** 2

grid = npa([[a,b,s,0] for a in lsa for b in lsb for s in lss])
for param in grid:
    likelihood = []
    for _ in range(0, 10):
        b = BFP("multnorm", K = 100, T = obs.size, alpha = alpha, beta = beta, sigma2 = sigma2, inference=True, obs=obs)
        likelihood.append(b.log_likelihood )
    param[3] = np.mean(likelihood)
```

```python
grid = grid[np.flip(grid[:, 3].argsort())]
for i in range(0, min(grid.shape[0], 10)):
    print("{}. alpha:{:0.2f}, beta:{:0.2f}, sigma^2:{:0.2f}, log likelihood:{}".format(i+1, grid[i][0],grid[i][1],grid[i][2],grid[i][3]))
```
